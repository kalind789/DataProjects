```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
link = "https://www.nba.com/stats/players/boxscores?CF=PLAYER_NAME*E*Joki&SeasonType=Regular%20Season&dir=D&sort=PLAYER_NAME"
page = requests.get(link).text
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome()
driver.get(link)
soup = BeautifulSoup(driver.page_source, 'html.parser')
table = soup.find_all("tbody", class_="Crom_body__UYOcU")
```

```python
table = table[0]
```

```python
rows = table.find_all("tr")
len(rows)
```

```python
headers = soup.find_all("tr", class_="Crom_headers__mzI_m")
headers = headers[0]
```

```python
elements = headers.find_all("th")
elements
```

```python
col_names = []
for th in elements:
    col_names.append(th.get_text(strip=True))

print(col_names)
```

```python
stats = []
for row in rows:
    list = []
    for td in row.find_all("td"):
        a = td.find("a")
        if a in td:
            list.append(a.get_text(strip=True))
        else: 
            list.append(td.get_text(strip=True))
    stats.append(list)

print(stats)
print(len(stats))
```

```python
df = pd.DataFrame(stats, columns=col_names)
df.head()
```

```python
df.to_csv('jokic.csv')
```

```python
# A triple double is where a player gets more than 10 of each major stat
# Major stats include, PTS, REB, AST, STL, BLK

# Check triple double
def check_triple_double(row):
    count = 0
    for stat in ['PTS', 'AST', 'REB', 'BLK', 'STL']:
        try:
            if int(row[stat]) >= 10:
                count += 1
        except:
            continue  # Skip if stat is missing or not an int
    return count >= 3


df['Triple Double'] = df.apply(check_triple_double, axis=1)
df.head()
```

```python
def check_almost_triple_double(row):
    count = sum([
        int(row['PTS']) >= 9,
        int(row['AST']) >= 9,
        int(row['REB']) >= 9,
        int(row['BLK']) >= 9,
        int(row['STL']) >= 9,
    ])

    if check_triple_double(row) == False:
        return count >= 3
    else: 
        return False

df['Almost Triple Double'] = df.apply(check_almost_triple_double, axis=1)
```

```python
df_w_almost_triple_double = df[df['Almost Triple Double'] == True]
df_w_almost_triple_double
```

```python
def check_double_double(row):
    count = sum([
        int(row['PTS']) >= 10,
        int(row['AST']) >= 10,
        int(row['REB']) >= 10,
        int(row['BLK']) >= 10,
        int(row['STL']) >= 10,
    ])
    return count >= 2
```

```python
jokic_df['Double Double'] = jokic_df.apply(check_double_double, axis=1)
jokic_df['Double Double'].sum() # True = 1, False = 0 so sum of True = total
```

```python
def check_triple_double(row):
    count = 0
    for stat in ['PTS', 'AST', 'REB', 'BLK', 'STL']:
        try:
            if int(row[stat]) >= 10:
                count += 1
        except:
            continue  # Skip if stat is missing or not an int
    return count >= 3

def check_almost_triple_double(row):
    count = sum([
        int(row['PTS']) >= 9,
        int(row['AST']) >= 9,
        int(row['REB']) >= 9,
        int(row['BLK']) >= 9,
        int(row['STL']) >= 9,
    ])

    if check_triple_double(row) == False:
        return count >= 3
    else: 
        return False

def get_player_triple_double(name, season="2024-25"):
    player_id = players.find_players_by_full_name(name)[0]['id']
    player_df = playergamelogs.PlayerGameLogs(player_id_nullable=player_id, season_nullable=season).get_data_frames()[0]

    return player_df.apply(check_triple_double, axis=1).sum()

def get_player_almost_triple_double(name, season="2024-25"):
    player_id = players.find_players_by_full_name(name)[0]['id']
    player_df = playergamelogs.PlayerGameLogs(player_id_nullable=player_id, season_nullable=season).get_data_frames()[0]

    return player_df.apply(check_almost_triple_double, axis=1).sum()

print("'almost' triple doubles means 10+ 10+ 9")
print("Westbrook (2016-17) had ", get_player_triple_double(name="Russell Westbrook", season="2016-17"), " triple doubles.")
print("Westbrook had ", get_player_almost_triple_double(name="Russell Westbrook", season="2016-17"), " almost triple doubles.")

print("Jokic (2024-25) had ", get_player_triple_double(name="Nikola Jokic", season="2024-25"), " triple doubles.")
print("Jokic had ", get_player_almost_triple_double(name="Nikola Jokic", season="2024-25"), " almost triple doubles.")

print("Westbrook played 81 games, Jokic played 70")
```
