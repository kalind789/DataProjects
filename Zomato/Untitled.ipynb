```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato-data-.csv")
print(dataframe.head())
```

```python
def handleRate(value):
    value=value.split('/')
    value=value[0]
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Restaurant")
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
results = pd.DataFrame({'votes': grouped_data})
print(results.head())
plt.plot(results, c='g', marker='o')
plt.xlabel('Types of restaurants')
plt.ylabel('Votes')
```

```python
maxvotes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == maxvotes, 'name']

print('Restaurants with max votes: ' + restaurant_with_max_votes)
```

```python
sns.countplot(x=dataframe['online_order'])
```

```python
plt.hist(dataframe['rate'], bins=5)
plt.title('Ratings Distribution')
plt.show()
```

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order', y='rate', data=dataframe)
```
